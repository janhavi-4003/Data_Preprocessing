```python
#handling missing value (mean and scikit_leaarn)
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy='mean')
ds[['Age','Salary']]=si.fit_transform(ds[['Age','Salary']])
```

```python
# encoding catogerical colomn (OnehotEncoding)
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
encoded_data=ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
# make column headers and make a dataframe for array
column_name=ohe.get_feature_names_out(['Country'])
encoded_df=pd.DataFrame(encoded_data,columns=column_name)
encoded_df
```

```python
# Joining original data and new dataframe
ds=pd.concat([ds,encoded_df], axis=1)
ds
```

```python
# droping categorical column
ds=ds.drop(['Country'], axis=1)
ds
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
#feature scaling(Reduce larger data values to its smaller data values)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ds[['Age','Salary']]=sc.fit_transform(ds[['Age','Salary',]])
```

```python
#Divide dataset as Independent(input ) and Dependent(output) variables

x=ds[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y=ds[['Purchased']]
```
